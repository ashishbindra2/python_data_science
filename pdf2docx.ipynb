```python
import pypandoc

def convert_docx_to_pdf(input_path, output_path):
    output = pypandoc.convert_file(input_path, 'pdf', outputfile=output_path, extra_args=['--pdf-engine=wkhtmltopdf'])
    assert output == "", "An error occurred during conversion"
    print(f"File converted successfully and saved to {output_path}")

# Example usage
input_docx = 'Holidays.docx'  # Replace with your DOCX file path
output_pdf = 'Hos.pdf'   # Replace with your desired PDF file path

convert_docx_to_pdf(input_docx, output_pdf)
```

```python
import os
import pypandoc
from weasyprint import HTML

def convert_docx_to_pdf(input_path, output_path):
    # Convert DOCX to HTML
    html_output = pypandoc.convert_file(input_path, 'html')

    # Save the HTML content to a temporary file
    html_file = input_path.replace('.docx', '.html')
    with open(html_file, 'w', encoding='utf-8') as f:
        f.write(html_output)
    
    # Convert HTML to PDF using WeasyPrint
    HTML(html_file).write_pdf(output_path)
    os.remove(html_file)  # Clean up the temporary HTML file

    print(f"File converted successfully and saved to {output_path}")

# Example usage
try:
    convert_docx_to_pdf("Holidays.docx", "abbb.pdf")
except Exception as e:
    print(f"An error occurred: {e}")
```
