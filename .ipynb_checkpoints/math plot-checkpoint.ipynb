2. Relative Frequency

```python
import matplotlib.pyplot as plt

# Data
x_axies = ["Beach", "City", "Adventure", "Nature", "Cruise", "Other"]
relative_frequency = [0.3, 0.2, 0.15, 0.175, 0.1, 0.075]

# Create pie chart
plt.pie(relative_frequency, labels=x_axies, autopct='%1.1f%%',  textprops=dict(color="w"))

plt.title("Relative Frequency")

plt.legend(x_axies,
          title="Types of Vacation",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel file
df = pd.read_excel('./train.xlsx')

# Extract relevant columns
survived = df["Survived"]
pclass = df["Pclass"]

# Create a crosstab to get the count of each Pclass within Survived categories
crosstab = pd.crosstab(survived, pclass)

# Plot the data
crosstab.plot(kind='bar', figsize=(10, 7))

plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Count by Passenger Class')
plt.legend(title='Pclass')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel file
df = pd.read_excel('./train.xlsx')

# Extract relevant columns
survived = df["Survived"]
pclass = df["Pclass"]

# Create a crosstab to get the count of each Pclass within Survived categories
crosstab = pd.crosstab(survived, pclass)

# Get the number of unique Pclass values
pclass_values = crosstab.columns

# Define the width of the bars
bar_width = 0.35

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 7))

# Plot bars for each Pclass
for i, pclass_value in enumerate(pclass_values):
    # Calculate the bottom for stacking
    if i == 0:
        ax.bar(crosstab.index, crosstab[pclass_value], label=f'Pclass {pclass_value}')
    else:
        bottom = crosstab[pclass_values[:i]].sum(axis=1)
        ax.bar(crosstab.index, crosstab[pclass_value], bottom=bottom, label=f'Pclass {pclass_value}')

# Add labels and title
ax.set_xlabel('Survived')
ax.set_ylabel('Count')
ax.set_title('Survival Count by Passenger Class')

# Add legend
ax.legend(title='Pclass')

# Show plot
plt.show()
```
