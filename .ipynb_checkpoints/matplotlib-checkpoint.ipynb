```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,11)
y = x**2
```

```python
plt.plot(x,y) # (1,1),(2,4),(3,9)......
plt.show()
```

```python
plt.plot(x,y) # (1,1),(2,4),(3,9)......
plt.title('square function line plot')
plt.show()
```

```python
plt.plot(x,y) # (1,1),(2,4),(3,9)......
plt.title('square function line plot')
plt.xlabel('N value')
plt.ylabel('square of N')
plt.show()
```

- Line style property

**Line Styles**
    
    =============    ===============================
    character        description
    =============    ===============================

```python
plt.plot(x,y, linestyle='. .')# if we try to use any other style we will get Value error
plt.show() 
```

### multiple lines on the same plot

```python
x = np.arange(1,11)
i = x
s = x **2
c = x **3
```

```python
plt.plot(x,i,'o-r')
plt.plot(x,s,'o-b')
plt.plot(x,c,'o-g')
plt.title('One Grapth, but multiple plot')
```

### customization of xlabel and ylabel

```python
a = np.arange(1,11)
b = a**2
```

```python
plt.plot(a,b)
plt.title('sqluare functionline plot')
plt.xlabel('year',{'color':'r','size':20,'backgroundcolor':'yellow',
                   'rotation':10,'alpha':1,'fontstyle':'italic',
                   'family':'cursive','weight':1000})
```

```python
plt.plot(a,b)
plt.title('sqluare functionline plot')
plt.xlabel('year',{'color':'r','size':20,'backgroundcolor':'yellow',
                   'rotation':10,'alpha':1,'fontstyle':'italic',
                   'fontfamily':'cursive','weight':1000})
plt.ylabel('data Science',color='r',size=20,backgroundcolor='yellow',
                   rotation=100,alpha=1,fontstyle='italic',
                   fontfamily='cursive',weight=1000)
```

```python
help(plt.title)
```

```python
plt.plot(a,b)
plt.title('sqluare functionline plot')
plt.grid(color='g')
# plt.grid()
```

```python
plt.plot(a,b)
plt.title('sqluare functionline plot')
plt.grid(visible=False)
```

```python
plt.plot(a,b)
plt.title('sqluare functionline plot')
plt.minorticks_on()
plt.grid(which='minor')
```

```python
plt.plot(a,b)
plt.title('sqluare functionline plot')
plt.minorticks_on()
plt.grid(which='both')
```

```python
plt.plot(a,a,'o-r',lw=7 , markersize=10,mfc='yellow')
plt.grid(color='r', linewidth=3)
plt.title('sqluare functionline plot')
plt.minorticks_on()
plt.grid(which='minor',color='g')
```
