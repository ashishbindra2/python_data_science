```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1,11)
y = x**2
```

```python
plt.plot(x,y) # (1,1),(2,4),(3,9)......
plt.show()
```

```python
plt.plot(x,y) # (1,1),(2,4),(3,9)......
plt.title('square function line plot')
plt.show()
```

```python
plt.plot(x,y) # (1,1),(2,4),(3,9)......
plt.title('square function line plot')
plt.xlabel('N value')
plt.ylabel('square of N')
plt.show()
```

## line properties
1. Marker property

```python
plt.plot(x,y,marker = 'o') # (1,1),(2,4),(3,9)......
plt.show()
```

- Line style property

**Line Styles**
    
    =============    ===============================
    character        description
    =============    ===============================
    ``'-'``          solid line style
    ``'--'``         dashed line style
    ``'-.'``         dash-dot line style
    ``':'``          dotted line style
    =============    ===============================

```python
plt.plot(x,y, linestyle=':')
plt.show()
```

```python
plt.plot(x,y, linestyle='-')
plt.show()
```

```python
plt.plot(x,y, linestyle='--')
plt.show()
```

```python
plt.plot(x,y, linestyle='-.')
plt.show()
```

```python
plt.plot(x,y, linestyle='. .')# if we try to use any other style we will get Value error
plt.show() 
```

3. Color prperty
 **Colors**
    
    The supported color abbreviations are the single letter codes
    
    =============    ===============================
    character        color
    =============    ===============================
    ``'b'``          blue
    ``'g'``          green
    ``'r'``          red
    ``'c'``          cyan
    ``'m'``          magenta
    ``'y'``          yellow
    ``'k'``          black
    ``'w'``          white
    =============    ===============================
    
    and the ``'CN'`` colors that index into the default property cycle.
    
    If the color is the only part of the format string, you can
    additionally use any  `matplotlib.colors` spec, e.g. full names
    (``'green'``) or hex strings (``'#008000'``).

```python
plt.plot(x,y,color='r')
plt.show()
```

# shortcut way to specify all three marker,line style,color property

```python
plt.plot(x,y,marker='o',linestyle='-',color='r') # without shortcut way
```

```python
plt.plot(x,y,'o-r') # mlc
```

```python
plt.plot(x,y,'ro-') # cml
```

```python
plt.plot(x,y,'o-red')# invalid
```
