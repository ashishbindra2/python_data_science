### Importing a CSV into Python


```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

```

### Importing a CSV without a Header
- read_csv() to read csv file
- read_csv assume data contains header


```python
df = pd.read_csv(url,header=None) # not assine headers
```

### Printing the dataframe in python
- ***df*** prints the entire dataframe (Not recommended for large datasets)
- ***df.head(n)*** to show the first n rows of data frame
- ***df.tails(n)*** to show the bottom n rows of data frame

```python
df.head()
```

### Adding headers
- Replace default header (by `df.columns = headers`)
- It is diffcult to work with the dataframe without meaningful column names
- but we can assing them


```python
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style", "drive-wheels",
           "engine-location", "wheel-base","length","width","height", "curb-weight", "engine-type", "num-of-cylinders",
           "engine-size","fuel-system","bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg",
           "highway-mpg","price"]
df.columns = headers
```
