### Why check data types

- Potential infi and type mismatch
    - pandas automatically assings types based on encoding it detects from orignal data table
- Compatibilty with python methods
    - some math function only need numerical data if these functions are applied to non numerical data an error may occure

#### Check type
- In pandas, we use `dataframe.dtpes` to check data types
    - eg `df.dtypes`

```python
df.dtypes
```

### Statistical summary
#### dataframe.describe()
- Returns a statistical summary
- `df.describe()` => To learn about the distribution of data ineach column
- The statistical metrics can tell the data scientist if there are mathematical issues that may exist
    - such as extreme outliers and large deviations

```python
df.describe()
```

### dataframe.describe(include = "all")
- Provide full summary statistics
- `df.describe(include="all")`
- it include object type attributes
- top : most time occure
- freq: number of time top occurs
- NAN : it is NOT A NUMBER

```python
df.describe(include="all")
```
