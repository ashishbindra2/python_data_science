### Data Normalization
- Uniform the features value with different range
- Normalization make some statistical analyses easier.
- By making range consistent between variables Normalization enables a faire comparison between features making sure they have same impact
- it also important to computational resons


### 
- age ranges 0-100 where income ranges 0 to 200000
- income is about 1000 times larger than age


### Linear regration
- in linear regration income will influence the result more due to its large value.
- Nature of the data baiases the linear regression model to weigh income more heavily than age.
- we can normalize these two variables into values that range from 0 to 1


### Methods of Normalizing data
- Several approaches for normalization:

1. Simple Feature scaling (value range 0-1)
2. Min-Max (value range 0-1)
3. Z-score (standard score) (value range -3-3)



### Simple Feature Scaling 


```python
import pandas as pd
import numpy as np

headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style", "drive-wheels",
           "engine-location", "wheel-base","length","width","height", "curb-weight", "engine-type", "num-of-cylinders",
           "engine-size","fuel-system","bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg",
           "highway-mpg","price"]

df = pd.read_csv("../data/automobile.csv")

df.columns = headers

df.head()
```

```python
df["length"] = df["length"] / df["length"].max()
df["length"]
```

### Min-max
- on length feature


```python
df["length"] = (df["length"] - df["length"].min()) / (df["length"].max() - df["length"].min())
df["length"]
```

### Z-score
U (mue) = average of feature
sigma () = statndard devation


```python
df["length"] = (df["length"] - df["length"].mean()) / df["length"].std()
df["length"]
```
