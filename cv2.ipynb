```python
import cv2
import pytesseract
import os
from pytesseract import Output

os.environ['TESSDATA_PREFIX'] = r'C:\Program Files\Tesseract-OCR\tessdata'

# Load the image using OpenCV
image_path = "./20220831072149_00030.tif"
image_cv = cv2.imread(image_path)

# Perform OCR on the image to get bounding box data
data = pytesseract.image_to_data(image_cv, output_type=Output.DICT)

# Extract words and their coordinates
words = []
for i in range(len(data['level'])):
    word = data['text'][i]
    if word.strip():
        x, y, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
        words.append((word, x, y, w, h))

# Draw rectangles around each detected word
for word, x, y, w, h in words:
    # Draw a rectangle with a border color of (0, 255, 0) which is green
    cv2.rectangle(image_cv, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Save the annotated image
annotated_image_path = "./annotated_image.tif"
cv2.imshow('Annotated Image', image_cv)
cv2.waitKey(0) 

cv2.imwrite(annotated_image_path, image_cv)
```

```python
import cv2
import pytesseract
from pytesseract import Output
import matplotlib.pyplot as plt

# Load the image using OpenCV
image_path = "1.png"
image_cv = cv2.imread(image_path)

# Perform OCR on the image to get bounding box data
data = pytesseract.image_to_data(image_cv ,output_type=Output.DICT)
print(data)
# Extract words and their coordinates
words = []
for i in range(len(data['level'])):
    word = data['text'][i]
    if word.strip():
        x, y, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
        words.append((word, x, y, w, h))

# Draw rectangles around each detected word
for word, x, y, w, h in words:
    # Draw a rectangle with a border color of (0, 255, 0) which is green
    cv2.rectangle(image_cv, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Convert the image from BGR to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)

# Display the annotated image using matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# Save the annotated image
annotated_image_path = "annotated_image.tif"
cv2.imwrite(annotated_image_path, image_cv)
```

```python
import cv2
import easyocr
import matplotlib.pyplot as plt

# Load the image using OpenCV
image_path = "20220831072149_00030.tif"
image_cv = cv2.imread(image_path)

# Initialize EasyOCR Reader
reader = easyocr.Reader(['en'])

# Perform OCR on the image
results = reader.readtext(image_path)

# Draw rectangles around each detected word
for (bbox, text, prob) in results:
    # Unpack the bounding box coordinates
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = tuple(map(int, top_left))
    bottom_right = tuple(map(int, bottom_right))

    # Draw a rectangle around the word
    cv2.rectangle(image_cv, top_left, bottom_right, (0, 255, 0), 2)

# Convert the image from BGR to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB)

# Display the annotated image using matplotlib
plt.figure(figsize=(15, 15))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# Save the annotated image
annotated_image_path = "annotated_image.tif"
cv2.imwrite(annotated_image_path, image_cv)
```
