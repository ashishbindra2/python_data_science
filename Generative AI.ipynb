#### Synthetic data 
Synthetic data is data that has been created artificially through computer simulation or that algorithms can generate to take the place of real-world data.
-  One example is synthetic data used in healthcare to protect patient data and enhance clinical trials
- The technique is gaining in popularity in the further development of deep learning and many other use cases. 
- 60% of the data used in training AI models will be synthetically generated.


#### How does synthetic data work?
Synthetic data is created programmatically with machine learning techniques to mirror the statistical properties of real-world data. Synthetic data can be generated in a multitude of ways, with really no limit to size, time, or location.
